```python
# Two Sum:

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        store = dict()
        for i in range(len(nums)):
            sec = target - nums[i]
            if sec not in store:
                store[nums[i]]=i
            else:
                return [store[sec],i] 
```

```python
# Write a program to find the square root of a double

def find_root(n):
    start = 0
    end = n

    while((end - start)>=0.000001):

        mid = (start + end)/2;

        if(mid*mid < n):
            start = mid;
        if(mid*mid >= n):
            end = mid;
    return start

find_root(10)
    
```

```python
# code a fibonacci
def Fibonacci(n):

    if n < 0:
        print("Incorrect input")
 
    elif n == 0:
        return 0
 
    elif n == 1 or n == 2:
        return 1
 
    else:
        return Fibonacci(n-1) + Fibonacci(n-2)
 
print(Fibonacci(9))
```

```python
# Count of possible seating arrangements in Cinema hall to maintain Social Distancing
# https://www.geeksforgeeks.org/count-of-possible-seating-arrangements-in-cinema-hall-to-maintain-social-distancing/

Use a dp[] array to store the possible seating arrangements up to the ith seat. 
Check each of the seat starting from 1st. 
If the distance between current seat ‘i’ and previous seat ‘j’ is valid (>= 6), add dp[j] to dp[i]

def get_possible_seatings(seats):  
    dist = 6
     
    # Account for the last seat
    seats.append(0)
 
    # Each seat can be occupied individually
    dp = [1] * len(seats)
 
    # Keep track of total distance
    # from first seat
    total_distance = [0] * len(seats) # total distabce from the dirst seat 
    prefix_sum = seats[0]
    for index, i in enumerate(seats[1:], 1):
        total_distance[index] = prefix_sum
        prefix_sum += i
 
    # Start from second seat onwards,
    # this is the curr seat 'i'
    for i in range(1, len(seats)):
        for j in range(i):
            if total_distance[i] \
            - total_distance[j] >= dist:
                dp[i] += dp[j]
 
    # Account for no seat occupied
    return sum(dp) + 1
 
# Driver code
if __name__ == "__main__":
    list = [5, 2, 4, 1, 2]
    ans = get_possible_seatings(list)
    print(ans)
    
```
